````python
from dotenv import load_dotenv

load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
assistant_messages = []

def add_user_message(messages, text):
    user_message = ({"role": "user", "content": text})
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = ({"role": "assistant", "content": text})
    messages.append(assistant_message)
    ''
def chat(messages, system=None, temperature=1.0, stop_sequences=["5","3, 4"]):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }
    
    if system is not None:
        params["system"] = system
        
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
# make a starting list of messages
messages = []
system = """
You are a python engineering assistant.
When answering the question provide code that is as concise as possible. Suggest only one solution to the problem.
Only provide the code, nothing else.
"""

add_user_message(messages, "Count from 1 to 10")
answer = chat(messages=messages)
answer
````
